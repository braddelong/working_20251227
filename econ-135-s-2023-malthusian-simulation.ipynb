```python
import numpy as np               # math package
import matplotlib.pyplot as plt  # plotting package
import random as rnd             # random number package
```

```python
g = []                           # time-series list for productivity growth
n = []                           # time-series list for population growth
epsilon = []                     # uncorrelated random shocks to population
L = [1]                          # time-series list for population & labor force level
y = [1.25]                       # time-series list for productivity & living standard level

h = 0.0005                       # annual rate of technological progress
gamma = 2                        # salience of ideas vis-Ã -vis resources in determining productivity
beta = 0.025                     # annual response of population growth to relative living standards

for t in range(999):                                        # loop through time for 1000 years
    epsilon = epsilon + [rnd.normalvariate(0, 0.02)]        # calculate this year's population shock
    n = n + [beta * (y[t] - 1) + epsilon[t]]                # calculate this year's population growth rate
    g = g + [h - n[t]/gamma + rnd.normalvariate(0, 0.0005)] # calculate this year's productivity change
    y = y + [y[t]*np.exp(g[t])]                             # calculate this year's productivity level
    L = L + [L[t]*np.exp(n[t])]                             # calculate this year's population level
```

```python
tseries = range(1000)      # set up a time-series list of years for the horizontal axis


fig, ax = plt.subplots()   # open a portal for graphs

ax.plot(tseries, L, y)     # plot living standards and population over time
```

```python
fig, ax = plt.subplots()   # open a portal for graphs

ax.plot(L, y)              # plot living standards against population            
```
