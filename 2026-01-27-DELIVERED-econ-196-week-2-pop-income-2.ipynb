### A Word About This "pop[-73000] = 0.01" Millions—That Is, 10,000 People
<p style="text-align: right;">

* The gene-science people will tell us this:
* It really looks like the overwhelming majority of us derive the overwhelming proportion of our ancestry from about 100 groups of 100 or so wandering around East-Central Africa some 75000 years ago.
* But what if we ask them a different question: how many hominins alive 75000 years ago have living descendants today?
* They will run and hide...
* Pin them down, and they will say genes are "flowing' within and between;
    * East-African _homines sapientes sapientes_ population (100,000?)
    * South-African _homines sapientes sapientes_ population (Khoesan: 75,000?)
    * Eurasian Central Asia and further west _homines sapientes neandertalenses_ population (75000?)
    * Eurasian Central Asia and further east _homines sapientes denisovenses_ population (50,000?)
    * African & Eurasian "ghost hominins" (100,000)
    * For a total of 400,000 as a **VERY ROUGH AND NOT WELL MOTIVATED GUESS**
    * By 50,000 years ago, primarily East-African _homines sapientes sapientes_ lineages have spread throughout Africa and are spreading over Eurasia, and the entire population of us _homines sapientes_ are shifting to primarily their genes (but with admixtures)
* Thus you could say: pop[-73000] = 0.5
* That is the group of people who have left living descendants today, and their cousins—all those whom are ancestors found attractive enough to mate with and raise children with.

```python
# time to look at growth rates

import numpy as np

# 1. Make sure there is no index/column ambiguity and you're sorted by year
human_history_df = human_history_df.reset_index(drop=True)
human_history_df = human_history_df.sort_values("year")

# 2. Year gaps
human_history_df["year_diff"] = human_history_df["year"].diff()

# 3. Annual *log* growth rates
human_history_df["pop_growth"] = (
    np.log(human_history_df["population"])
    - np.log(human_history_df["population"].shift(1))
) / human_history_df["year_diff"]

human_history_df["income_growth"] = (
    np.log(human_history_df["income"])
    - np.log(human_history_df["income"].shift(1))
) / human_history_df["year_diff"]

human_history_df
```

```python
# calculating "technology"—growth rate equal to income growth plus half of population growth

human_history_df["tech_growth"] = (
    human_history_df["income_growth"] + 0.5 * human_history_df["pop_growth"]
)

human_history_df
```

```python
#constructing a level index for "technology"

human_history_df = human_history_df.sort_values("year").reset_index(drop=True)

human_history_df["ln_tech"] = np.nan

human_history_df.loc[human_history_df["year"] == -73000, "ln_tech"] = 0.0

# build ln_tech forward from the first non-NaN
base_idx = human_history_df["ln_tech"].first_valid_index()
mask_after_base = human_history_df.index > base_idx

human_history_df.loc[mask_after_base, "ln_tech"] = (
    human_history_df.loc[mask_after_base, "tech_growth"]
    * human_history_df.loc[mask_after_base, "year_diff"]
).cumsum()

human_history_df
```

```python
# 4. exponentiate to get a tech index (base = 1 at year -73000)

human_history_df["tech_index"] = np.exp(human_history_df["ln_tech"])

# find the tech_index value at 1875
base_value_1875 = human_history_df.loc[human_history_df["year"] == 1875, "tech_index"].iloc[0]

# rescale the entire index so that tech_index(1875) = 1
human_history_df["tech_index"] = human_history_df["tech_index"] / base_value_1875

human_history_df
```

```python
# population, income, growth rate hockey-stick

ax = human_history_df.plot(
    x="year",
    y=["pop_growth", "income_growth", "tech_growth"],
    kind="line",
    legend=True
)

ax.set_title("Annual Log Growth Rates: Population, Income, Technology")
ax.set_xlabel("Year")
ax.set_ylabel("Annual log growth rate")

plt.tight_layout()
plt.show()
```

# The Demographic Transition

Title: The Demographic Transition: Three Centuries of Fundamental Change    
Author: Ronald Lee    
JEP 17(4), 2003

* Core question: How did the mortality–fertility transition reshape population size, age structure, and economic life over the past three centuries?
* Key claim: Since ~1800, declining mortality followed by declining fertility has transformed population growth, age distributions, family behavior, and the economic lifecycle.

---
---

## The Demographic Transition in Three Moves

* Phase 1 – Mortality decline: Starting ~1800 in (north‑western) Europe, mortality—especially infant and child mortality—falls sharply, initially with little change in fertility.
* Phase 2 – Population surge: With high fertility and lower mortality, population size explodes and the age structure becomes very young; dependency initially falls because working‑age cohorts are abundant.
* Phase 3 – Fertility decline & aging: Fertility later falls toward replacement or below, population growth slows or turns negative, and age structures shift from young to old, producing population aging. •⁠•

---
---

## Drivers of Mortality Decline

* Public health & environment: Clean water, sanitation, and better control of infectious disease (smallpox, later TB, etc.) cut mortality well before “modern medicine” becomes effective.
* Nutrition and living standards: Rising incomes and better food distribution improve resistance to disease, interacting with public health to lower death rates.
* Medical progress later: Vaccines, antibiotics, and clinical medicine become quantitatively important later in the transition, reinforcing earlier gains.
* Result: Big mortality improvements at all ages, but especially infants and children, set off the demographic transition. •⁠•

---
---

## Drivers of Fertility Decline

* Economic calculus of children: As child survival improves, parents need fewer births to achieve desired surviving children; the “insurance” motive weakens.
* Rising costs of children: Urbanization, schooling, and foregone child labor raise the direct and opportunity costs of each child.
* Changing norms & contraception: Ideational change (family limitation becoming legitimate) and access to birth control technologies reduce fertility at given income levels.
* Gender & education: Rising female education and labor‑force participation are strongly associated with lower desired fertility and later childbearing. •⁠•

---
---

## From Young to Old Society: Changing Age Structures

* Early transition – youth bulge: When mortality falls and fertility remains high, child shares of the population rise and dependency can initially increase.
* Demographic dividend window: As fertility falls, large cohorts move into working ages; the ratio of workers to dependents rises, potentially boosting per capita income growth if employment and institutions are supportive.
* Late transition – aging: Continued low fertility plus sustained low mortality at older ages shift the population toward the elderly, raising old‑age dependency and straining pay‑as‑you‑go pension and health systems. •⁠•

---
---

## Individual & Family-Level Consequences

* Longer lives, longer horizons: Individuals must plan over longer lifespans, changing savings, education, and retirement behavior.
* Fewer children, more investment per child: Parents substitute “quality” for “quantity”—spending more on each child’s education and health.
* Changing intergenerational transfers: The shape of the economic lifecycle shifts; time and resources flow differently across generations, often mediated by public systems (pensions, schooling, health insurance) rather than only by families. •⁠•

---
---

## Macro Consequences: Growth, Capital, & the Welfare State

* Growth and the demographic dividend: A favorable age structure (large working‑age cohorts) can raise aggregate saving, labor supply, and growth, especially in the “East Asian miracle” cases.
* Capital accumulation & saving: With fewer children and longer retirement, life‑cycle saving behavior changes; in some phases this may raise aggregate saving and capital per worker.
* Population aging & fiscal stress: Low fertility and longevity together create rising old‑age support ratios, pressuring pensions, health systems, and intergenerational fiscal contracts; policy responses (later retirement, higher contributions, benefit cuts, immigration) become central. •⁠•

---
---

## Lee’s Bottom Line

* Fundamental transformation: The demographic transition is not a marginal adjustment but one of the great structural changes of the last three centuries, comparable in significance to industrialization.
* Dual nature: It is both a success story (longer lives, smaller families, scope for more human capital) and a challenge (population aging, pension and health‑care burdens, potential for slower growth).
* Open future: The process is incomplete—many developing countries are still moving through the transition, and advanced economies are grappling with below‑replacement fertility and rapid aging. •⁠•

---
---

## DeLong’s Take I: Demographic Transition as the End of Human Ensorcellment by the Devil of Malthus

* Pre‑1870 world as Malthusian: For most of human history, technological improvements fed primarily into more people, not higher sustained living standards; mortality and fertility jointly kept societies “ensorcelled by the Devil of Malthus.”
* Post‑1870 “Economic El Dorado”: After 1870, rapid technological progress and the demographic transition together allow sustained growth in income per capita; technology finally “wins its race against fecundity.” •⁠•
* Lee as the missing chapter: In his own words, DeLong concedes that Slouching Towards Utopia pushed the demographic transition “offstage,” treating it as a simple mechanical response to higher incomes and lower infant mortality—an “inadequate treatment” of one of the most important pieces of twentieth‑century history. •⁠•

---
---

## DeLong’s Take II: Power, Gender, & Today’s Low-Fertility World

* Beyond income–fertility mechanics: DeLong is suspicious of any story in which demographic transition is just an Oded‑Galor‑style shift from quantity to quality of children; he insists on the role of increased female social power and political struggle—“my great‑grandmother and all my sisters did this”—in driving fertility decline. ​⁠￼
* Demography and the unfinished utopia: In his grand narrative, the twentieth century reveals that we can bake a big enough economic pie; the fact that we have not solved distribution and “living wisely and well” is the core failure. The demographic transition is central here: it both creates the possibility of sustained abundance and, via aging and the political economy of welfare states, generates new conflicts over “who gets what, when, and how.” •⁠•
* Policy implication in DeLong’s register: For rich countries facing low fertility and rapid aging, the challenge is to use the wealth enabled by technology and the demographic transition to build institutions—around migration, family policy, gender equality, pensions, and health—that keep the social contract from cracking, rather than letting demographic anxiety fuel zero‑sum politics.

# Bill Nordhaus on the Price of Artificial Illumination, & the Puzzle of the True Rate of Economic Growth

**Central question**: Do standard real GDP and real wage measures capture the true rise in living standards, or are they missing the big welfare gains from technological revolutions like artificial illumination? 

### Nordhaus’s Object Lesson: From Tallow to Tungsten (and Beyond)

* Service, not good: Nordhaus prices illumination (lumen‑hours), not fuel or hardware, over two centuries.
* Technologies sequenced: Tallow candles → whale oil → gas lighting → kerosene → incandescent bulbs → fluorescents, etc.
* Order‑of‑magnitude fall: The cost of a lumen‑hour falls by factors of thousands; a given amount of work time buys orders of magnitude more light in 1990 than in 1800.
* Implication: Measured prices of inputs (candles, oil, electricity) largely miss the productivity explosion in the service of light. ​⁠https://lucept.com/wp-content/uploads/2014/11/william-nordhaus-the-cost-of-light.pdf

---
---

## Why Standard Real-Output Measures Miss the Story

* Laspeyres problem: Official price indices follow input prices in a narrow bundle, holding quality and product definition fixed; they rarely re‑specify the good as “lumens” rather than “candles” or “kWh.”
* Quality change at regime breaks: Each major lighting regime shift (gas → kerosene → electricity) brings huge jumps in efficiency, reliability, and usability that standard indices treat as small or gradual.
* Result: Traditional real-output and real-wage measures drastically understate the decline in the “true” price of light and, hence, understate growth in real living standards. ​⁠https://lucept.com/wp-content/uploads/2014/11/william-nordhaus-the-cost-of-light.pdf

---
---

##Nordhaus’s Broader Claim about Economic Growth

* Lighting as a microcosm: Illumination is a case study for a pervasive phenomenon: when technologies leap, measured prices lag.
* Invisible revolutions: Similar mismeasurement likely for heating, transport, communication, information, and health—domains where the service improved more than the thing whose price we track.
* Timing of fastest gains: For light, Nordhaus finds the largest unmeasured productivity jump roughly in the half century around the late nineteenth century; the “special century” looks even more special once you correct for this.
* Takeaway: True long‑run growth in economic welfare is substantially higher than what national accounts report. •⁠https://lucept.com/wp-content/uploads/2014/11/william-nordhaus-the-cost-of-light.pdf

---
---

## DeLong on “How Fast Is Modern Economic Growth?”

* Official story: Post‑Civil War U.S. productivity growth ~1.6%/year ⇒ output per worker doubles every ~44 years; three doublings since 1867. ​⁠https://www.frbsf.org/research-and-insights/publications/economic-letter/1998/10/how-fast-is-modern-economic-growth/
* Historical comparison: Even 1770–1867 plausibly saw one doubling; 1867–today sees three by official measures. Already enormous by pre‑modern standards.
* But…: Evidence has been mounting that this already “blazingly fast” growth is understated, for exactly Nordhaus‑type reasons: we misprice new goods, new qualities, and entirely new activities.
* FRBSF hook: The Economic Letter explicitly uses Nordhaus’s illumination work as a canonical example of how standard productivity measures miss big welfare gains. <<https://www.frbsf.org/research-and-insights/publications/economic-letter/1998/10/how-fast-is-modern-economic-growth/>>

---
---

## DeLong’s “True Rate” View: Off the Scale of 19th-Century Imagination

* Bellamy thought experiment: In Looking Backward, being able to dial up live music into your parlor would have been “the limit of human felicity”; modern recorded music and streaming services go well beyond that imagined ceiling. ​⁠https://www.frbsf.org/research-and-insights/publications/economic-letter/1998/10/how-fast-is-modern-economic-growth/
* Richest then vs. average now: Even the Morgans and Rothschilds lacked antibiotics, central heating, modern entertainment, telephones, IT; they simply could not buy today’s consumption bundle at any price.
* Conclusion: From the vantage point of 1867, the material standard of living of an average American today sits “off the scale” of what they could even conceptualize, implying true growth far exceeds the 8x increase in measured GDP per worker. •⁠https://www.frbsf.org/research-and-insights/publications/economic-letter/1998/10/how-fast-is-modern-economic-growth/
* But distribution matters: Even if the average household has seen off‑the‑scale gains, those at the bottom who still spend most of their budget on basic food, shelter, and clothing may not have captured anything like that multiple; mismeasurement is larger higher up the income distribution. •⁠https://www.frbsf.org/research-and-insights/publications/economic-letter/1998/10/how-fast-is-modern-economic-growth/

---
---

## How Nordhaus’s Light Fits into DeLong’s Big Story

* Lighting as emblem of the 1870–1910 break: Nordhaus’s finding that the price of illumination collapses around the turn of the twentieth century dovetails perfectly with DeLong’s claim that a true 1870 watershed sees the world escape the Malthusian devil. ​⁠https://www.frbsf.org/research-and-insights/publications/economic-letter/1998/10/how-fast-is-modern-economic-growth/
* From “goods” to “capabilities”: Both Nordhaus and DeLong argue that we ought to measure growth in terms of what people can do—see at night, listen to music, travel, avoid infection—rather than the nominal value of the physical objects they buy.
* Shared bottom line: The official productivity series already show an unprecedented acceleration after 1870; once you fold in Nordhaus‑style service measures and DeLong’s concerns about new goods and consumer surplus, the true rate of growth of human technological capability is higher still.

---
---

## DeLong’s Assessment: What the Price of Light Teaches Us

* On measurement: Nordhaus demonstrates that even in a seemingly simple domain like lighting, conventional real-output and real-wage measures dramatically understate progress; extend that logic across all major technologies and modern measured growth is a lower bound, not a best estimate.
* On periodization: The magnitude and timing of the lighting revolution strengthen the case for a “long twentieth century” beginning around 1870, with a discrete jump in the underlying rate of technological change and welfare growth. ​⁠https://braddelong.substack.com/p/brad-delongs-history-of-economic
* On welfare and policy: Recognizing that true growth is faster than the national accounts say does not make distributive conflict or stagnation anxieties go away—if anything, it sharpens them: the pie has grown even more than we thought, and yet its slicing (and the treatment of environmental and social externalities) remains our central failure.

- William D. Nordhaus, “Do Real-Output and Real-Wage Measures Capture Reality? The History of Lighting Suggests Not,” in The Economics of New Goods (1997), argues that standard real-output and real-wage measures substantially understate true growth.

- Conventional measures show roughly a 15‑fold increase in rich‑country GDP per capita and productivity since ~1800, but Nordhaus’s lighting-based calculations suggest a much larger effective rise in real wages—on the order of 20‑fold to 30,000‑fold, depending on how you measure welfare.

- Nordhaus decomposes 1991 consumption into three categories: about 28% “run‑of‑the‑mill” goods, 36% “seismically active” since 1800 (huge quality/price shifts), and 37% categories that had no affordable counterpart in 1800, raising deep questions (e.g., via Dixit‑Stiglitz) about how to aggregate utility across old and new varieties.

- Using illumination as a case study, Nordhaus finds about a 5,000‑fold fall in the price of light since 1800; because lighting had taken 1–5% of household budgets in 1800, a 100‑fold bias in the CPI for light accumulates into a major understatement of long‑run growth.

- The piano/music example illustrates service vs. good pricing: the labor‑time cost of a Steinway fell from ~2,400 worker‑hours a century ago to ~1,100 today, but the labor‑time cost of the capability to listen to music at home fell from ~2,400 hours to perhaps ~10 hours (streaming via a smartphone), implying roughly a 240‑fold gain in access to that capability.

- Edward Bellamy’s Looking Backward is used to make the welfare point vivid: for a late‑19th‑century utopian, simply dialing one of four live orchestras on a speakerphone would have represented “the limit of human felicity,” yet modern consumers treat far superior, on‑demand, effectively infinite recorded music as mundane.

- Extending Nordhaus’s logic, long‑run service‑based price indices would show much faster growth than commodity‑based ones; Nordhaus’s illumination series implies about a 10,000‑fold decline in the real labor‑time cost of light over the past century, versus only a 10‑fold decline captured by standard indices.

- Nordhaus’s best guess is that standard statistics understate “true” economic growth since 1800 by roughly 0.5–1.4 percentage points per year, implying that real wages since 1900 may have risen by a factor between 20 and 100 rather than by the much smaller multiples implied by official data.

- A time‑travel thought experiment reinforces credibility: even placed in the outer elite circle of 1890 with an income 64× median, a modern upper‑middle‑class family would feel poor because they would lack three crucial service bundles we now take for granted—modern health care, utility infrastructure and appliances, and information/communication/entertainment technologies.

- The death of Nathan Mayer Rothschild (from an abscess curable today with a few dollars’ worth of antibiotics), the crippling of FDR by polio, and the absence of modern communications all underscore the asymmetry: in many life‑and‑death or quality‑of‑life dimensions, even the richest in 1890 could not purchase what an average household now receives as part of everyday life, which is precisely the kind of welfare gain standard GDP and wage measures fail to capture.
